```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["font.size"]=15
plt.figure(dpi=120)

phi = np.linspace(0,1,20)
p1 = phi**3*(10-15*phi+6*phi**2)
p2 = phi**2*(3-2*phi)
plt.plot(phi,p1,'-o',color='b',label='p^3(10-15p+6p^2)')
plt.plot(phi,p2,'-o',color='r',label='p^2(3-2p)')

plt.xlim(0,1)
plt.ylim(0,1)
plt.xlabel('phase-field')
plt.ylabel('interpolation function')
plt.legend(fontsize=13)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["font.size"]=15
plt.figure(dpi=120)

phi = np.linspace(0,1,30)
q = phi**2*(1-phi)**2
plt.plot(phi,q,'-o',color='black',label='p^2(1-p)^2')

plt.xlim(0,1)
plt.ylim(0,0.07)
plt.xlabel('phase-field')
plt.ylabel('double well potential function')
plt.legend(fontsize=13)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["font.size"]=15
plt.figure(dpi=120)

c = np.linspace(0,1,30)
f = c**2*(1-c)**2
plt.plot(c,f,'-o',color='black',label='c^2(1-c)^2')

plt.xlim(0,1)
plt.ylim(0,0.07)
plt.xlabel('concentration')
plt.ylabel('chemical free energy function')
plt.legend(fontsize=13)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["font.size"]=15
plt.figure(dpi=120)

ax = plt.subplot(111,projection="polar")
x = np.linspace(0,2*np.pi,100)
y = 1+0.06*np.cos(4*x)

ax.set_ylim([0,1.2])
ax.set_rgrids([0,0.5,1.0],labels=[])
ax.plot(x,y)
plt.show()
```
